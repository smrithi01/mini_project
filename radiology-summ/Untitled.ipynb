```python
from rouge import Rouge
import re

def evaluate_summaries(reference_file, generated_file):
    # Read the reference summaries from the file
    with open(reference_file, 'r') as f:
        reference_summaries = f.readlines()
    reference_summaries = [re.sub(r'\n', '', s) for s in reference_summaries]

    # Read the generated summaries from the file
    with open(generated_file, 'r') as f:
        generated_summaries = f.readlines()
    generated_summaries = [re.sub(r'\n', '', s) for s in generated_summaries]

    # Initialize Rouge
    rouge = Rouge()

    # Calculate ROUGE scores
    scores = []
    for hyp, ref in zip(generated_summaries, reference_summaries):
        try:
            score = rouge.get_scores([hyp], [ref])
            scores.append(score)
            print(f"ROUGE Scores for Generated Summary: {hyp}\nReference Summary: {ref}")
            print(score)
            print()
        except Exception as e:
            print(f"Error calculating ROUGE score for hypothesis '{hyp}' and reference '{ref}'")
            scores.append(None)

    # Check if scores is empty
    if not scores:
        print("Error: ROUGE scores are empty.")
        return None

    # Calculate average ROUGE scores
    avg_scores = {k: sum(v) / len(v) for k, v in zip(scores[0][0].keys(), zip(*[s[0].values() for s in scores]))}

    # Print the average ROUGE scores
    print("Average ROUGE Scores:")
    print(avg_scores)

    return avg_scores

# Example usage:
reference_file = "reference_summary.txt"
generated_file = "summaries.txt"
evaluate_summaries(reference_file, generated_file)
```
